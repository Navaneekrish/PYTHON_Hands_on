MATPLOTLIB

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
arr1 = np.arange(1,50,5)
arr1
```

```python
arr2=np.log(arr1)
arr2
```

```python
#To plot the graph
plt.plot(arr1,arr2,'b-.')
plt.xlabel('Numbers in X axis')
plt.ylabel('log in y axis')
plt.title('Logarithmic graph')
```

```python
#To create the two different graph of same one use subplot(row,column,graph name)
plt.subplot(1,2,1)
plt.plot(arr1,arr2,'g--')
plt.xlabel('Numbers in X axis')
plt.ylabel('log in y axis')

plt.subplot(1,2,2)
plt.plot(arr2,arr1,'r-.')
plt.ylabel('Numbers in y axis')
plt.xlabel('log in x axis')
```

```python
plt.subplot(2,2,1)
plt.plot(arr1,arr2,'g--')
plt.subplot(2,2,2)
plt.plot(arr2,arr1,'b-.')
plt.subplot(2,2,3)
plt.plot(arr1,arr2,'y--')
plt.subplot(2,2,4)
plt.plot(arr2,arr1,'r-.')
```

Object Oriented approach

```python
fig = plt.figure() #empty figure
#To create the own graph parameters(lenght,b,width,height)
axes = fig.add_axes([0.1,0.1,0.9,0.9])
axes.plot(arr1,arr2,'g--')
axes.set_xlabel('Number in x-axis')
axes.set_ylabel('log in y-axis')
axes.set_title('Logarithmic Graph')
```

```python
#Inline Graph
fig = plt.figure() #empty figure

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.55,0.21,0.4,0.4])
#gp 1
axes1.plot(arr1,arr2,'g--')
axes1.set_xlabel('Number in x-axis')
axes1.set_ylabel('log in y-axis')
axes1.set_title('Logarithmic Graph')
#gh 2
axes2.plot(arr2,arr1,'r--')
axes2.set_ylabel('Number in x-axis')
axes2.set_xlabel('log in y-axis')
axes2.set_title('Logarithmic Graph')
```

```python
#creating the figure and axes automatically using Subplots()
fig, axes = plt.subplots()

axes.plot(arr1,arr2,'g--')
axes.set_xlabel('Number in x-axis')
axes.set_ylabel('log in y-axis')
axes.set_title('Logarithmic Graph')
```

```python
#multi graph in subplots()
fig,axes = plt.subplots(nrows=1,ncols=2)
#gp 1
axes[0].plot(arr1,arr2,'g--')
#gh 2
axes[1].plot(arr2,arr1,'r--')
```

```python
#multi graph in subplots()
fig,axes = plt.subplots(nrows=2,ncols=2)
#gp 1
axes[0,0].plot(arr1,arr2,'g--')
#gh 2
axes[0,1].plot(arr2,arr1,'r--')
#gp 3
axes[1,0].plot(arr1,arr2,'y--')
#gh 4
axes[1,1].plot(arr2,arr1,'b--')
```

```python
#To enlarge the figure size use figsize()
fig = plt.figure(figsize=(6,6),dpi=100)
#dpi is The resolution of the figure in dots-per-inch.
#figsize  is  Width, height in inches.
axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.55,0.21,0.4,0.4])
#gp 1
axes1.plot(arr1,arr2,'g--')
axes1.set_xlabel('Number in x-axis')
axes1.set_ylabel('log in y-axis')
axes1.set_title('Logarithmic Graph')
#gh 2
axes2.plot(arr2,arr1,'r--')
axes2.set_ylabel('Number in x-axis')
axes2.set_xlabel('log in y-axis')
axes2.set_title('Logarithmic Graph')
```

```python
#To save the graph use savefig
fig.savefig('log_graph.png')
```

```python
#For legend by using label and for color and alpha(is color transparent)
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes1.plot(arr1,arr2,color='#808000',label='arr1 to arr2',alpha=0.3)
axes1.plot(arr2,arr1,color='#00FF00',label='arr2 to arr1')
plt.legend(loc=5)
```

```python
#Boxplot
fig,axes = plt.subplots(nrows=1,ncols=2,figsize=(9,5))

box1 = axes[0].boxplot(arr1,vert=True,notch=True,patch_artist=True)
patch,color = list(zip(box1['boxes'],['pink']))[0]
patch.set_facecolor(color)
axes[0].set_title('Notch Boxplot')

box2 = axes[1].boxplot(arr2,vert=True,patch_artist=True)
patch,color=list(zip(box2['boxes'],['Green']))[0]  #for custom color
patch.set_facecolor(color)
axes[1].set_title(' Boxplot')
```
