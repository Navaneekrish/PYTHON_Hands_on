```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

LINE GRAPH

```python
gas = pd.read_csv(r'D:\Class\Data collection\gas_prices.csv')
```

```python
gas
```

```python
#To increase the graph size
plt.figure(figsize=(8,8))
#To name the title
plt.title('Gas Price variation over Year')
plt.xlabel('Year')
plt.ylabel('Price in Dollar')
#To plot the graph
#'[marker][line][color]' in third place
plt.plot(gas.Year,gas['South Korea'],'b.-',label='S_Korea')
plt.plot(gas.Year,gas.UK,'+--r',label='United Kingdom',)
plt.plot(gas.Year,gas.Japan,'*:k',label='Japan',)
#To show the legends
plt.legend()
#To change the x axis to porper values
plt.xticks(gas.Year[::3])
plt.show()
```

```python
#To have all the country values in the same step
plt.figure(figsize=(8,8))
for country in  gas:
    if country != 'Year':
        plt.plot(gas.Year,gas[country],marker = '*')

#if u what to expand the year upto 2011 
plt.xticks(gas.Year[::3].tolist()+[2011])

#To show the legend of country which u selected 
plt.legend(gas,labels=['Australia', 'Canada', 'France', 'Germany', 'Italy', 'Japan',
       'Mexico', 'South Korea', 'UK', 'USA'])
     
```

```python
#if u using only selected countries in the visual then 
plt.title('Gas price in USD',fontdict={'fontsize':12,'fontweight':'bold'})
select_country =['Mexico', 'South Korea', 'UK', 'USA']

for country in gas:
    if country in select_country: 
        plt.plot(gas.Year,gas[country],marker='*')
#To arrange the Year in X axis use Xticks
plt.xticks(gas.Year[::3])

plt.legend(gas,labels=['Mexico', 'South Korea', 'UK', 'USA'])

#To save the graph as the image to ur document the use savefig

plt.savefig('Gas_price_Analysis.png',dpi=300)
```

```python
#Pie chart indivison to show 
short = fifa[fifa['Height']<="5'5"].count()[0]
medium_short = fifa[(fifa['Height']>"5'5") & (fifa['Height']<="6'0")].count()[0]
Height = fifa[(fifa['Height']>"6'0") & (fifa['Height']<="6'5")].count()[0]
Over_Height = fifa[(fifa['Height']>"6'5") & (fifa['Height']<="7'0")].count()[0]
```

```python
#To view all the style
plt.style.available
```

```python
data = [short,medium_short,Height,Over_Height]
plt.style.use('bmh')  #for style
label = ["<5'5","<5'5 & <6'0","<6'0 & <6'5","<6'5 & <7'0"]
color = ['#74a955','#8a55a9','#cfad21','#2143cf']
explode = [.1,0,.1,.1] #to create some space
plt.pie(data,labels=labels,colors=color,explode=explode,autopct='%.2f %%')
plt.show()
```

```python
fifa.Weight= [int(x.strip('lbs')) if type(x) == str else x for x in fifa.Weight]
```

```python
fifa.Weight
```

```python
#Pie chart indivison to show 
light_lbs = fifa[fifa['Weight']<150].count()[0]
medium_lbs = fifa[(fifa['Weight']>=150) & (fifa['Weight']<200)].count()[0]
heavy_lbs = fifa[fifa['Weight']>=200].count()[0]
```

```python
lbs=[light_lbs,medium_lbs,heavy_lbs]
plt.style.use('tableau-colorblind10')
labels = ['light_lbs(<150)','medium_lbs(<200)','heavy_lbs(>200)']
explode = [0,.1,.1]
plt.pie(lbs,labels=labels,autopct='%.2f %%',explode=explode)
plt.show()
```
