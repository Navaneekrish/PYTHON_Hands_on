```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Sample')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y,'r')
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(x,y,'r')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.2,0.5,0.4,0.4])
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(x,y,'r')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_xlim([0,100])
axes1.set_ylim([0,200])
axes1.set_title('Zoom_Out')
axes2 = fig.add_axes([0.2,0.5,0.2,0.2])
axes2.plot(x,y,'r')
axes2.set_xlim([20,22])
axes2.set_ylim([30,50])
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Zoom_In')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2 )
axes[0].plot(x,y,'b',lw = 3,ls ='--')
axes[1].plot(x,y,'r',lw = 3,ls ='-')
```

```python
fig,axes = plt.subplots(nrows = 1,ncols = 2,figsize = (12,2))
axes[0].plot(x,y,'b',lw = 3,ls ='--')
axes[1].plot(x,y,'r',lw = 3,ls ='-')
```

```python
violinplot
A violin plot plays a similar role as a box and whisker plot. 
It shows the distribution of quantitative data across several levels of one (or more) categorical variables such that 
those distributions can be compared. Unlike a box plot, in which all of the plot components correspond to actual datapoints, 
the violin plot features a kernel density estimation of the underlying distribution.
```

```python
sns.violinplot(x='time',y='size',data=df,palette='rainbow')
```

```python
sns.violinplot(x ='day',y ='total_bill',data =df,hue ='sex')
```

```python
sns.violinplot(x ='size',y ='tip',data =df,hue ='sex',split =True)
```

```python
sns.stripplot(x='size',y ='tip',data =df)
```

```python
sns.stripplot(x='size',y ='tip',data =df,hue ='sex',jitter=True,palette='nipy_spectral')
```

```python
sns.swarmplot(x ='size',y ='tip',data =df,hue ='sex')
```
