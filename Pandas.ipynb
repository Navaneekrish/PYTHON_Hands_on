```python
#To slice the column from the table
df9=df8.loc['movie_5':,'star_rating':'genre']
df9
```

```python
#Appling the condition filter onj the table values
df10 = df.loc[df['star_rating']>=9.0]
df10.sort_values('star_rating')
```

```python
#For implict index as 0,,1,2,3,4,5,6,.... use iloc[]
df11 = df.iloc[4:8]
df11
```

```python
#if we need all the columns but only three rows specific
df12 = df.iloc[[2,5,7],:]
df12
```

```python
#for filtering the specific value from the columns
df13=df[df['genre']=='Crime']
df13
```

```python
#for filtering more than one specific value from the columns use &
df14=df[(df['genre']=='Crime') & (df['star_rating']>9.0)]
df14
```

```python
#filtering either value from the columns |
df15=df[(df['genre']=='Crime') | (df['star_rating']>9.0)]
df15
```

```python
#for not equal to != or ~
df16=df[(df['genre']!='Crime')]
df16
```
