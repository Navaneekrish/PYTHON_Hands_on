Adjusting Date Ranges in Python as per the WeekDays or Holidays

```python
ss['Order Date'].nunique()
```

```python
#To take Business date which is except saturday and sunday use('B')
pd.date_range(start='2015-03-07',end='2015-05-01',freq='B')
```

```python
from  pandas.tseries.offsets import CustomBusinessDay
```

```python
#If we dont want to add the custom leave date then use (CustomBusinessDay)
cust_bday = CustomBusinessDay(holidays = ['2015-04-10'])
```

```python
pd.date_range(start='2015-03-07',end='2015-05-01',freq=cust_bday)
```

```python
pd.date_range(start='2015-03-07',end='2015-05-01',freq=cust_bday).nunique()
```

```python
#if u want to set calendar date as per US calendar Holidays then 
from pandas.tseries.holiday import USFederalHolidayCalendar
us_holiday = CustomBusinessDay(calendar = USFederalHolidayCalendar())
```

```python
pd.date_range(start='2015-03-07',end='2015-05-01',freq=us_holiday)
```

```python
pd.date_range(start='2015-03-07',end='2015-05-01',freq=us_holiday).nunique()
```

```python
#Different weekdays in Dubai(weekmask)
dubai_weekdays = CustomBusinessDay(weekmask = 'Sat Sun Mon Tue Wed',holidays=['2015-03-10'])
```

```python
pd.date_range(start='2015-03-07',end='2015-05-01',freq=dubai_weekdays)
```
