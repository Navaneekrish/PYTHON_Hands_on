```python
#Ploting distplot

sns.distplot(pokemon['Attack'])
plt.show()
```

```python
#For custom y axis use kde =false

hist = sns.distplot(pokemon['Attack'],kde = False)
#kde : bool, optional Whether to plot a gaussian kernel density estimate.
hist.set_xlabel('Attack points')
hist.set_ylabel('Frequency')
hist.set_title('Attack capability')
plt.show()
```

```python
#For only line chart use hist =False

hist = sns.distplot(pokemon['Attack'],hist = False)

hist.set_xlabel('Attack points')
hist.set_title('Attack capability')
plt.show()
```

```python
#For rug graph in histogram 

hist = sns.distplot(pokemon['Attack'],rug = True)

hist.set_xlabel('Attack points')
hist.set_ylabel('Frequency')
hist.set_title('Attack capability')
plt.show()
```

```python
#Rugplot graph
rug=sns.rugplot(pokemon['Attack'],height=0.5 ,axis='x')
rug.set_xlabel('Attack points')
rug.set_ylabel('Frequency')
rug.set_title('Attack capability')
plt.show()
```

```python
#distplot and rug without line graph
dist=sns.distplot(pokemon['Attack'],bins=100 ,kde = False,rug=True)
#bins :Specification of hist bins. If unspecified, as reference rule is used that tries to find a useful default.
dist.set_xlabel('Attack points')
dist.set_ylabel('Frequency')
dist.set_title('Attack capability')
plt.show()
```

```python
#line chart with shade inside
sns.kdeplot(pokemon['Attack'],color='g',shade=True)
plt.show()
```

Bivariate Data
The graph display inside is bivariate data (in default 'scatter plot') use 'kind' to change it.

```python
sns.jointplot(data=pokemon,x='Attack',y='Defense') # we can also custom the xlim & ylim
plt.show()
```

```python
#kind : { "scatter" | "kde" | "hist" | "hex" | "reg" | "resid" }
#Kind of plot to draw. See the examples for references to the underlying functions.
#To change the axes background color use axes_style()
with sns.axes_style('white'):
    sns.jointplot(data=pokemon,x='Attack',
              y='Defense',kind='hex',
              xlim=(200,0),ylim=(0,200),
             color='r') # we can also custom the xlim & ylim
plt.show()
```

```python
#Kind = 'kde'
with sns.axes_style('white'):
    sns.jointplot(data=pokemon,x='Attack',shade=True,
              y='Defense',kind='kde',
             color='r')
plt.show()
```
